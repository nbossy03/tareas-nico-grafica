Problema 2.

Hacer un programa que imprima los números del -100 al 200 que sean divisibles entre 6 (con resto 0)

Esto signifca que si el número es 6, 12, 18, etc. entonces debe imprimirlo ya que la división de esos números por 6 da como resto 0, exacto.

6 / 6 = 1, resto 0
12 / 6 = 2, resto 0
Distinto de:

7 / 6 = 1, resto 1
Hint de ayuda: en programación, hay un símbolo que ocupamos para saber inmediatamente el resto, en lugar de hacer la división y luego ver el resto. En python es el símbolo %, por ejemplo:

4 % 3 = 1
5 % 3 = 2
6 % 3 = 0 ✅ nos sirve (y deberíamos imprimirlo)

```python
# problema 3 de la tarea 02
numero1 = int(input("Ingrese el primer número: "))
numero2 = int(input("Ingrese el segundo número: "))

suma = numero1 + numero2

if suma < 200:
    resultado = "menor a 200"
elif suma > 200 and suma < 250:
    resultado = "mayor a 200 y menor a 250"
else:
    resultado = "es mayor a 250"

print(resultado)
```

```python
# problema 4 de la tarea 03
def evaluar_usuario(edad, gusta_musica_urbana):
  if edad > 18 and gusta_musica_urbana:
     resultado = "Eres mayor de edad y te gusta la música urbana"
  elif edad > 18 and not gusta_musica_urbana:
       resultado = "Eres mayor de edad y no te gusta la música urbana"
  elif edad <= 18 and gusta_musica_urbana:
      resultado = "Eres menor de edad y te gusta la música urbana"
  else:
     resultado = "Eres menor de edad y no te gusta la música urbana"
  return resultado
# Bloque principal del programa
try:
    edad = int(input("¿Cuál es tu edad?"))
    respuesta_musica = input("¿Te gusta la música urbana? (sí/no): ").strip().lower()

    if respuesta_musica == "sí" or respuesta_musica == "si":
        gusta_musica_urbana = True
    else:
        gusta_musica_urbana = False

    mensaje = evaluar_usuario(edad, gusta_musica_urbana)
    print(mensaje)

except ValueError:
    print("Por favor, introduce un número válido para la edad.")
```
